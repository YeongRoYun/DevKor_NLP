### 2-2. Numpy

   * #### ndarray
     * 행렬 표현의 기본!
     * 리스트, 튜플, 배열로 행렬을 생성한다
     * 몇 차원 배열을 생성하던, 타입은 동일하다(ndarray)
     * 축(ndim) : 행의 수(독립된 리스트의 수)
     * 크기(shape) : (ndim, col)

````python
import numpy as np

#리스트로 생성
vec = np.array([1,2,3,4,5])
print("리스트로 생성")
print("-"*15)
print(vec)
print("-"*15)
mat = np.array([[10, 20, 30], [40, 50, 60],])
print(mat)
print("-"*15)

print("vec의 타입 : ", type(vec))
print("mat의 타입 : ", type(mat))
print("-"*15)

print("vec의 축의 개수 : ", vec.ndim)
print("vec의 크기 : ", vec.shape)
print("mat의 축의 개수 : ", mat.ndim)
print("mat의 크기 : ", mat.shape)
print("-"*15)

#다른 방법
print("다양한 생성 방법")
print("-"*15)
zero_mat = np.zeros((2,3))
print(zero_mat)
print("-"*15)

one_mat = np.ones((2,3))
print(one_mat)
print("-"*15)

same_value_mat = np.full((2,2), 7)
print(same_value_mat)
print("-"*15)

eye_mat = np.eye(3)
print(eye_mat)
print("-"*15)

random_mat = np.random.random((2,2))
print(random_mat)
print("-"*15)

range_vec = np.arange(10)
print(range_vec)
print("-"*15)

#배열의 구조 변경
print("배열의 구조 변경")
print("-"*15)
reshape_mat = np.array(np.arange(30)).reshape((5,6))
print(reshape_mat)
print("-"*15)

mat = np.array([[1,2,3], [4,5,6]])
#슬라이싱
print("슬라이싱")
print("-"*15)
print(mat)
print("-"*15)
print("첫번째 행")
print("-"*15)
print(mat[0, :])
print("-"*15)
print("두번째 열")
print(mat[:,1])
print("-"*15)

#인덱싱
print("인덱싱")
print("-"*15)
print("값 하나 추출")
print("-"*15)
print(mat[1,0])
print("-"*15)
print("값을 추출해 새로운 배열")
print("-"*15)
print(mat[[1,1], [0,0]])
print("-"*15)

#연산
print("연산")
print("-"*15)
x = np.array([1,2,3])
y = np.array([4,5,6])
print("x")
print("-"*15)
print(x)
print("-"*15)
print("y")
print("-"*15)
print(y)
print("-"*15)

print("x + y")
print(x+y) #np.add(x,y)
print("-"*15)

print("x - y")
print(x-y) #np.subtract(x,y)
print("-"*15)

print("x * y(pairwise)")
print(x*y) #np.multiply(x,y)
print("-"*15)

print("x / y(pairwise)")
print(x/y) #np.divide(x,y)
print("-"*15)

print("<x,y>")
print(np.dot(x,y)) #행렬 곱
print("-"*15)

print("x X y")
print(np.cross(x,y)) #cross product
print("-"*15)

print(" x (X) y")
print(np.outer(x,y)) #outer product
````

## 05. 텍스트 전처리

### 5-1. 전처리 과정
- 코퍼스 데이터가 전처리되지 않은 상태일 때, 토큰화, 정제, 정규화 작업을 진행한다

### 5-2. 토큰화(Tokenization)
- #### 단어 토큰화(Word Tokenization)
  - 토큰의 기준: 단어, 단어구, 의미를 갖는 문자열
  
  - 기준에 대한 고려
    - Don't be fooled by the dark sounding name, Mr.Jone's Orphanage is as cheery as cheery goes for a pastry shop
    - Don't, Jone's를 어떻게 토큰화 할것인가?
      - Dont, Don t, Do n't, Jone, Jone's, ...
  
  - 토큰화 도구
    - NLTK
      - word_tokenize
      - WordPunctTokenizer
    - Keras
      - text_to_word_sequence
      
  - 고려사항
    - 구두점과 특수 문자를 단순히 제거하면 안된다
      - 구두점조차도 하나의 토큰으로 분류하기도 한다
        - '.'은 문장의 경계를 아는데 도움을 준다
        - Ph.D 등 단어 자체에 구두점이 있을 수 있다
        - 01/12/2022는 시간을 의미
      - 줄임말과 단어 내에 띄어쓰기가 있는 경우
        - 영어권의 " ' "는 압축된 단어를 펼치는 역할을 한다
          - what're == what are(re is clitic)
        - New York, rock'n'roll은 하나의 단어다
      - 표준 토큰화 예제
        - Penn Treebank Tokenization's rule
          - 규칙 1: '-'으로 구성된 단어는 하나로 유지
          - 규칙 2:'doesn't'처럼 아포스트로피로 '접어'가 함께하는 단어는 분리

- #### 문장 토큰화(Sentence Tokenization)
  - 토큰의 기준: 문장
  - 문장 분류(sentence segmentation)이라고도 부른다
  - 코퍼스가 정제되지 않았을 때, 문장 단위로 구분되어 있지 않을 가능성이 크다
  - 기준에 대한 고려
    - ?,!는 꽤 명확한 구분자(boundary)
    - .는 불분명할 수 있다
      - IP 192.168.0.3으로 보내주세요
   - 토큰화 도구
     - NLTK
       - sent_tokenize
     - KSS(Korean Sentence Splitter)
- #### 한국어에서 토큰화의 어려움
  - 영어의 경우, 'New York', 'he's' 같은 합성어와 줄임말에 대한 예외처리를 하면 띄어쓰기(whitespace) 단어 토큰화도 잘 동작한다
  - 한국어는 띄어쓰기로 토큰화하기 힘들다
    - 어절(띄어쓰기로 구분) 토큰화 != 단어 토큰화
    - 한국어는 교착어다
  - 교착어
    - 조사, 어미 등을 어근에 붙여 말을 만드는 언어
    - 그가, 그에게, 그를, 그와, 그는 등처럼 조사는 띄어쓰기 없이 바로 붙는다
    - 조사는 분리해줄 필요가 있다
  - 형태소(morpheme)
    - 의미를 가진 가장 작은 말의 단위
    - 종류
      - 자립 형태소
        - 독립적으로 사용할 수 있는 형태소
        - 체언(명사, 대명사, 수사), 수식언(관형사, 부사), 감탄사 등
      - 의존 형태소
        - 결합되어 사용해야 하는 형태소
        - 접사, 어미, 조사, 어간 등
    - 에디가 책을 읽었다
      - 띄어쓰기 토큰화: [에디가, 책을, 읽었다]
      - 형태소 토큰화: [에디, 가, 책, 을, 읽-, -었-, -다]
    - 영어의 단어 토큰화와 유사한 결과를 얻으려먼 형태소 토큰화를 해야 한다
   - 영어보다 띄어쓰기가 잘 지켜지지 않는다
     - 띄어쓰지 않아도 이해가능
     - 한국어는 모아쓰기 방식이고 영어는 풀어쓰기 방식이라는 언어적 특성에 기인한다
       - 모아쓰기: ㅂ + ㅏ + ㅂ = 밥
       - 풀어쓰기: r + i + c + e = rice
- #### 품사 태깅(Part-of-speech tagging)
  - 품사에 따라 의미가 달라지기도 한다
    - fly: v-날다, n-파리
  - 토큰화 과정에서 단어의 품사를 구분한다
  - 품사 태깅 도구
    - NLTK
      - 기준: Penn Treebank POS Tags
        - PRP: 인칭 대명사
        - VBP: 동사
        - RB: 부사
        - VBG: 현재부사
        - IN: 전치사
        - NNP: 고유명사
        - NNS: 복수형 명사
        - CC: 접속사
        - DT: 관사
    - KoNLPy
      - 형태소 및 품사 태깅 도구
      - Okt(Open Korea Text)
      - Mecab
      - Komoran
      - Hannanum
      - Kkma(꼬꼬마)
      - interface
        - morphs: 형태소 추출
        - pos: 품사 태깅
        - nouns: 명사 추출

### 5-3. 정제와 정규화(Cleaning&Normalization)
- 토큰화 작업 전후로 코파스를 정제 및 정규화해야 한다
- #### 의미
  - 정제
    - 코퍼스로부터 노이즈 데이터를 제거한다
  - 정규화
    - 표현 방법이 다른 단어들을 통합하여 같은 단어로 만든다
  - 어느정도의 타협점을 갖는다
- #### 대, 소문자 통합
  - 단어의 개수를 줄일 수 있는 정규화 방법
  - 소문자로 변환하는 작업으로 이루어진다
  - 주의사항
    - US와 us는 구분된다
    - 회사, 사람의 이름 등은 대문자로 유지되는 것이 옳다
  - 일부 문자만 소문자로 변환하는 방법
    - 문장의 첫 문자만 소문자로 변환
    - 머신러닝 시퀸스 모델로 정확하게 진행할 수 있다
  - 코퍼스가 소문자만 사용하는 데이터라면 그냥 전부 소문자로 바꾸는게 실용적일 수 있다
- #### 불필요한 단어의 제거
  - 아무 의미가 없거나 분석하는 목적에 맞지 않는 데이터
  - 방법
    - 불용어 제거
    - 등장 빈도가 적은 단어
    - 길이가 짧은 단어
      - 영어권에 해당

### 5-4. 어간 추출과 표제어 추출(Stemming&Lemmatization)
- 정규화 기법으로, 유의어를 하나의 단어로 만든다
- BoW 표현을 사용하는 자연어 처리 문제에서 주로 사용된다
- 차이점
  - am
    - Stemming: am
    - Lemmatization: be
  - the going
    - Stemming: the go
    - Lemmatization: the going
  - having
    - Stemming: hav
    - Lemmatization: have
- #### 어간 추출
  - 형태학적 분석을 단순화한 버전
  - 규칙만 보고 단어의 어미만 자른다
  - 추출의 결과가 사전에 등재되지 않을 수도 있다
  - 도구
    - NLTK
      - Porter Algorithm
      - Lancaster Stemmer Algorithm

- #### 표제어 추출
  - 단어의 원형을 찾는다
  - 가장 섬세한 방법은 형태학적 파싱을 먼저 진행
    - 어간(stem): 의미 함유
    - 접사(affix): 추가 의미 함유
    - 두 요소를 분리하는 작업
  - 도구
    - NLTK
      - WordNetLemmatizer
        - 단어의 품사 정보를 알아야 정확한 결과를 얻을 수 있다
          - lemmatizer.lemmatize('dies', 'v') #'die'
  - 표제어 추출은 문맥을 고려하며 수행했을 때, 단어의 품사 정보를 보존한다

- #### 한국어에서의 어간 추출
  - 한국어의 구조
  
  | 언 | 품사 |
  | :---: | :---: |
  | 체언 | 명사, 대명사, 수사 |
  | 수식언 | 관형사, 부사 |
  | 관계언 | 조사 |
  | 독립언 | 감탄사 |
  | 용언 | 동사, 형용사 |
  
  - 용언은 어간과 어미의 결합으로 구성된다
  - 활용(conjugation)
    - 인도유럽어족에서 주로 보이는 특징
    - 어간이 어미를 가지는 일
    - 불규칙 활용일 경우, 어간을 분리하기 힘들다
